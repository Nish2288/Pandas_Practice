```python
import pandas as pd
import datetime as dt
from pandas_datareader import data, wb
```

```python
start = dt.datetime(2015, 1, 1)
end = dt.datetime.today()
```

```python
end
```

## Select the Apple, Tesla,  IBM stocks symbols and assign them to a variable called stocks

```python
stocks = ['AAPL', 'TSLA', 'IBM']
```

## Read the data from google, assign to df and print it

```python
df = data.DataReader(stocks, 'yahoo', start, end)
```

```python
df.head()
```

## Create a dataFrame called vol, with the Volume values.¶

```python
volume = df['Volume']
volume.head()
```

## Aggregate the data of Volume to weekly.

```python
volume['week'] = volume.index.week
volume['year'] = volume.index.year
```

```python
volume.groupby(['week', 'year']).sum()
```

## Find all the volume traded in the year of 2015.

```python
volume[volume.year==2015].sum()
```
