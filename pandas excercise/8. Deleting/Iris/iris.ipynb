```python
import pandas as pd
import numpy as np
```

```python
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
iris = pd.read_csv(url)

iris.head()
```

## add columns to dataframe.

```python
iris.columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'class']
iris.head()
```

##  Is there any missing value in the dataframe?

```python
iris.isnull().sum()
```

## Lets set the values of the rows 10 to 29 of the column 'petal_length' to NaN

```python
iris.iloc[10:30, 2:3] = np.nan
```

```python
iris.head(30)
```

## now lets substitute the NaN values to 1.0

```python
iris.petal_length.fillna(1, inplace=True)
```

```python
iris.head(30)
```

## Now let's delete the column class

```python
del iris['class']
```

```python
iris.head()
```

##  Set the first 3 rows as NaN

```python
iris[0:3]=np.nan
```

```python
iris.head()
```

##  Delete the rows that have NaN

```python
iris = iris.dropna()
```

```python
iris.head()
```

## Reset the index so it begins with 0 again

```python
iris = iris.reset_index()
```
