```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("chipotle", sep = "\t")
```

```python
df.head()
```

```python
df.dtypes
```

## Convert item_price to float

```python
df["item_price"] = df["item_price"].map(lambda x : float(x[1:-1]))
```

```python
df.head(10)
```

```python
df.info()
```

## check for duplicates

```python
df[df.duplicated()]
```

## drop duplicates

```python
df_filtered = df.drop_duplicates(["item_name", "choice_description", "item_price"])
```

```python
df_filtered.info()
```

```python
df_filtered.head()
```

## select only the products with quantity equals to 1

```python
df_one_prod = df_filtered[df.quantity == 1]
df_one_prod
```

```python
df_one_prod[df_one_prod.item_price>10]
```

## delete the duplicates in item_name and quantity

```python
df.drop_duplicates(["item_name", "quantity"])
```

## # select only the products with quantity equals to 1

```python
df[df.quantity==1]
```

## select only the item_name and item_price columns

```python
df[["item_name", "item_price"]]
```

## sort the values from the most to less expensive

```python
df.sort_values(by="item_price", ascending=False)
```

##  Sort by the name of the item

```python
df.sort_values(by="item_name")
```

##  What was the quantity of the most expensive item ordered?

```python
df.sort_values(by="item_price", ascending=False).head(1)
```

## How many times was a Veggie Salad Bowl ordered?¶

```python
df[df.item_name == "Veggie Salad Bowl"].count()
```

##  How many times did someone order more than one Canned Soda?¶

```python
df[(df["item_name"] == "Canned Soda") & (df["quantity"]>1)].count()
```
