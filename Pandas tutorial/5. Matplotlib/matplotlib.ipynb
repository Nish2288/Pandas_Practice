```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
titanic=pd.read_csv('titanic.csv')
titanic.head()
```

## Numerical data

```python
titanic.plot()
plt.show()
```

```python
titanic.plot(subplots=True)
plt.show()
```

```python
titanic.plot(subplots=True, figsize=(15,12))
plt.show()
```

```python
titanic.age.plot(figsize=(12,8))
plt.show()
```

## Customization

```python
titanic.age.plot(figsize=(12,8), c='r',linestyle=":")
plt.title("Titanic- Age")
plt.xlabel("Index")
plt.ylabel("Age")
plt.legend()
plt.show()
```

## Histogram
Frequency distribution

```python
titanic.age.value_counts(bins=10)
```

```python
titanic.age.plot(kind="hist")
plt.show()
```

```python
titanic.age.plot(kind="hist", bins=80)
plt.show()
```

```python
# hist()
titanic.age.hist()
plt.show()
```

```python
# OR
plt.hist(titanic.age)
plt.show()
```
