# Filtering DataFrame with on condition

```python
import pandas as pd
```

```python
titanic = pd.read_csv('titanic.csv')
```

```python
# filter female
titanic.loc[titanic.sex=='female']
```

```python
titanic.loc[(titanic.sex=='female') & (titanic.age<10)].head()
```

```python
titanic.loc[(titanic.sex=='female') & (titanic.age<8)].describe()
```

```python
## another way: chain
titanic[(titanic.sex=='female')]['age']
```

## Filtering with between(), isin() and ~

```python
summer= pd.read_csv('summer.csv')
summer.head()
```

```python
#  Between()

summer[summer['Year'].between(1960,1969, inclusive=True)]
```

```python
## IN
yr = [1972, 1996]

summer.loc[summer.Year.isin(yr)]
```
