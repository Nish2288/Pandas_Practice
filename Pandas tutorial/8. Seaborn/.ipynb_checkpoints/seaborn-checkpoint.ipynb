```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
titanic = pd.read_csv("titanic.csv")
titanic.head()
```

## Categorical plot

```python
sns.catplot(x="sex", kind="count", data=titanic)
plt.show()
```

```python
sns.catplot(x="sex",hue="survived", kind="count", data=titanic)
```

```python
sns.catplot(x="sex",y='age', data=titanic)
plt.show()
```

```python
sns.catplot(x="sex",y='age', data=titanic, jitter=True,hue='survived', dodge=True, kind='strip')
plt.show()
```

```python
sns.catplot(x="sex",y='age', data=titanic,hue='survived', dodge=True, kind="swarm")
plt.show()
```

```python
sns.catplot(x="sex",y='age', data=titanic,hue='survived', dodge=True, kind="violin")
plt.show()
```

```python
sns.violinplot(x="sex",y='age', data=titanic,hue='survived', dodge=True, split=True)
plt.show()
```

```python
sns.catplot(x="pclass",y='fare', data=titanic,hue='survived', dodge=True, kind="bar")
plt.show()
```

```python
sns.catplot(x="pclass",y='fare', data=titanic,hue='survived', dodge=True, kind="point")
plt.show()
```

## Numerical columns

```python
sns.jointplot(x='age', y="fare", data=titanic)
```

```python
sns.jointplot(x='age', y="fare", data=titanic)
```

```python
sns.lmplot(x='age', y="fare", data=titanic)
```

```python
sns.lmplot(x='age', y="fare",hue="survived", data=titanic)
```

```python
sns.lmplot(x='age', y="fare",col="sex", data=titanic)
```

```python
sns.lmplot(x='age', y="survived", data=titanic, logistic=True)
```

```python
sns.lmplot(x='age', y="survived", data=titanic,col="sex", logistic=True)
```

```python
sns.lmplot(x='age', y="survived", data=titanic,col="pclass", logistic=True)
```

## Heatmap
Both categorical

```python
pd.crosstab(titanic.sex, titanic.pclass)
```

```python
plt.figure(figsize=(12,8))
sns.set(font_scale=1.4)
sns.heatmap(pd.crosstab(titanic.sex, titanic.pclass),annot=True,  fmt="d")
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.set(font_scale=1.4)
sns.heatmap(pd.crosstab(titanic.sex, titanic.pclass),annot=True,  fmt="d", vmax=150)
plt.show()
```

```python

pd.crosstab(titanic.sex, titanic.pclass,values=titanic.survived, aggfunc="mean")
```

```python
plt.figure(figsize=(12,8))
sns.set(font_scale=1.4)
sns.heatmap(pd.crosstab(titanic.sex, titanic.pclass,values=titanic.survived, aggfunc="mean"), annot=True)
```

```python
titanic.corr()
```

```python
plt.figure(figsize=(12,8))
sns.set(font_scale=1.4)
sns.heatmap(titanic.corr())
```
